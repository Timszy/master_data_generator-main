# HCO ground truth


```python
# Extract unique UUIDs from healthcare_organization_df_log
org_uuids = healthcare_organization_df_log['healthcareOrg_uuid'].unique()
address_uuids = healthcare_organization_df_log['address_uuid'].unique()
contact_uuids = healthcare_organization_df_log['contactPoint_uuid'].unique()
# Filter the changelog dataframes based on these UUIDs
filtered_org_df = healthcare_organization_df_changelog[
    healthcare_organization_df_changelog['original_id'].isin(org_uuids)
]
filtered_address_df = address_df_changelog[
    address_df_changelog['original_id'].isin(address_uuids)
]
filtered_contact_df = contactpoint_df_changelog[
    contactpoint_df_changelog['original_id'].isin(contact_uuids)
] 

# Concatenate the filtered dataframes
combined_entities_df = pd.concat(
    [filtered_org_df, filtered_address_df, filtered_contact_df], 
    ignore_index=False
)
combined_entities_df
```

```python
import re

omission_rows = []

# Iterate through the healthcare_organization_df_log to find omissions
for idx, log_row in healthcare_organization_df_log.iterrows():
    omitted_string = log_row['omitted']
    address_uuid = log_row['address_uuid']
    contact_uuid = log_row['contactPoint_uuid']
   

    # Parse address omissions
    address_match = re.search(r"address: ([\w\s,]+?)(?:, contact:|$)", omitted_string)
    
    if address_match:
        address_fields_str = address_match.group(1).strip()
        if address_fields_str:
            address_fields = [field.strip() for field in address_fields_str.split(',') if field.strip()]
           
            for field_name in address_fields:
                # Fetch original value from address_df_s
                original_value_series = address_df_s[address_df_s['identifier'] == address_uuid][field_name]
                original_value = original_value_series.iloc[0] if not original_value_series.empty else None
                
                
                omission_rows.append({
                    'row': idx,  # Using log_row index as the 'row' identifier
                    'original_id': address_uuid,
                    'field_name': field_name,
                    'original_value': original_value,
                    'varied_value': "",  # Empty as requested
                    'variation_type': 'omission',
                    'entity_type': 'Address'
                })

    # Parse contact point omissions
    contact_match = re.search(r"contact: \[(.*?)\]", omitted_string)
    if contact_match:
        contact_fields_str = contact_match.group(1).strip()
        if contact_fields_str:
            # Handles fields like 'email', 'phone', "'availableLanguage'"
            contact_fields = [field.strip().strip("'").strip('"') for field in contact_fields_str.split(',') if field.strip()]
            for field_name in contact_fields:
                # Fetch original value from contactpoint_df_s
                original_value_series = contactpoint_df_s[contactpoint_df_s['identifier'] == contact_uuid][field_name]
                original_value = original_value_series.iloc[0] if not original_value_series.empty else None
                
                
                omission_rows.append({
                    'row': idx, # Using log_row index as the 'row' identifier
                    'original_id': contact_uuid,
                    'field_name': field_name,
                    'original_value': original_value,
                    'varied_value': "",  # Empty as requested
                    'variation_type': 'omission',
                    'entity_type': 'ContactPoint'
                })

# Create a DataFrame from the omission rows
omissions_df = pd.DataFrame(omission_rows)

# Concatenate the new omission rows with the existing combined_entities_df
if not omissions_df.empty:
    combined_entities_df_om = pd.concat([combined_entities_df, omissions_df], ignore_index=True)

print("Shape of the updated combined_entities_df after adding omissions:", combined_entities_df_om.shape)
print("Few rows from the newly added omission data (if any):")
# Displaying rows that were just added (where variation_type is 'omission')
print(combined_entities_df_om[combined_entities_df_om['variation_type'] == 'omission'].head())
```

```python
import rdflib
from rdflib import Graph, Namespace, Literal, URIRef
from rdflib.namespace import RDF, RDFS, XSD
```

```python


# Create an RDF graph
g = Graph()

# Parse the Turtle file
g.parse('src/KGs/healthcareorganizations.ttl', format='turtle')
print(f"Successfully parsed TTL file. Graph contains {len(g)} triples.")
    
# Get a list of all namespaces used in the file
print("\nNamespaces in the file:")
for ns_prefix, namespace in g.namespaces():
    print(f"{ns_prefix}: {namespace}")
    
# Define namespaces that are commonly used
SCHEMA = Namespace("http://schema.org/")
EX = Namespace("http://example.org/")

```

```python
# Find all Address and ContactPoint entities in the RDF graph
address_entities = set()
contactpoint_entities = set()

for subj in g.subjects():
    subj_str = str(subj)
    if subj_str.startswith("http://example.org/Address/"):
        address_entities.add(subj)
    elif subj_str.startswith("http://example.org/ContactPoint/"):
        contactpoint_entities.add(subj)


print(contactpoint_entities)
#print("Address entities:", sorted(address_entities))
#print("ContactPoint entities:", sorted(contactpoint_entities))

for sub in address_entities:
    #print(f"Address entity: {sub}")
    for pred in g.predicate_objects(sub):
        #print (f"  Predicate: {pred[0]}")
        if str(pred[0]) == "https://schema.org/addressLocality":
            #print(f"Address entity: {sub}, addressLocality: {pred[1]}")
            if str(pred[1]) in address_df_noisy['city'].values:
                #print(f"Match found in address_df_noisy for addressLocality: {pred[1]}")
                #print(str(sub.split('/')[-1]), str(pred[1]))
                identifier = address_df_noisy.loc[address_df_noisy['city'] == str(pred[1]), 'identifier'].values[0]
                #print(f"Identifier: {identifier}") 
                if identifier in combined_entities_df_om['original_id'].values:
                    # Find the row in combined_entities_df_om with this original_id and update duplicate_id
                    mask = combined_entities_df_om['original_id'] == identifier
                    combined_entities_df_om.loc[mask, 'duplicate_id'] = str(sub.split('/')[-1])
                   
                    #print(f"Address entity: {sub}, addressLocality: {pred[1]}, Identifier: {identifier}")


for sub in contactpoint_entities:
    #print(f"ContactPoint entity: {sub}")
    for pred in g.predicate_objects(sub):
        #print (f"  Predicate: {pred[0]}")
        if str(pred[0]) == "https://schema.org/telephone" or str(pred[0]) == "https://schema.org/email":
            #print(f"ContactPoint entity: {sub}, availableLanguage: {pred[1]}")
            value = str(pred[1])
            if value.endswith('.'):
                value = value[:-1]
            if value in contactpoint_df_noisy['phone'].values:
                matching_rows = contactpoint_df_noisy.loc[contactpoint_df_noisy['phone'] == value, 'identifier']
                if not matching_rows.empty:
                    identifier = matching_rows.values[0]
                    if identifier in combined_entities_df_om['original_id'].values:
                        mask = combined_entities_df_om['original_id'] == identifier
                        combined_entities_df_om.loc[mask, 'duplicate_id'] = str(sub).split('/')[-1]
            
            # Check for email matches with safer approach
            elif value in contactpoint_df_noisy['email'].values:
                matching_rows = contactpoint_df_noisy.loc[contactpoint_df_noisy['email'] == value, 'identifier']
                if not matching_rows.empty:
                    identifier = matching_rows.values[0]
                    if identifier in combined_entities_df_om['original_id'].values:
                        mask = combined_entities_df_om['original_id'] == identifier
                        combined_entities_df_om.loc[mask, 'duplicate_id'] = str(sub).split('/')[-1]


# print("Address entity to addressLocality mapping:")
# for k, v in address_locality_map.items():
#     print(f"{k}: {v}")
# # For each address entity in address_df_noisy, match the addressLocality value with the city value
# # Assuming address_df_noisy has columns: 'identifier', 'addressLocality', 'city'
# if 'addressLocality' in address_df_noisy.columns and 'city' in address_df_noisy.columns:
#     address_df_noisy['addressLocality_matches_city'] = address_df_noisy['addressLocality'] == address_df_noisy['city']
#     print(address_df_noisy[['identifier', 'addressLocality', 'city', 'addressLocality_matches_city']])
# else:
#     print("addressLocality or city column not found in address_df_noisy")
```
